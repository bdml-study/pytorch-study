```python
from utils.data import CatsAndDogs
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
from settings import IMAGE_SIZE
```

```python
BATCH_SIZE = 8
```

```python
dataset = CatsAndDogs(mode="train")
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python
len(dataset)
```

```python
x, y = next(iter(dataloader))
```

```python
x.size(), y.size()
```

```python
x_np = x.numpy().reshape(BATCH_SIZE, IMAGE_SIZE, IMAGE_SIZE, 3) + 0.5
```

```python
for i in range(BATCH_SIZE):
    plt.imshow(x_np[i])
    plt.title(dataset.labels[int(y[i])])
    plt.xticks([], [])
    plt.yticks([], [])
    plt.show()
```
