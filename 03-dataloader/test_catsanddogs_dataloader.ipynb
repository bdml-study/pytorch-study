```python
from utils.catsanddogs_dataloader import CatsAndDogsDataLoader
import matplotlib.pyplot as plt
import numpy as np
```

```python
mydataloader = CatsAndDogsDataLoader(batch_size=8)
```

```python
for x, y in mydataloader:
    images, labels = x, y
    break
```

```python
images = images.astype(np.uint8)
labels = labels.astype(np.int32)
```

```python
for i in range(8):
    plt.imshow(images[i])
    plt.xticks([], [])
    plt.yticks([], [])
    plt.title(mydataloader.labels[labels[i]])
    plt.show()
```
