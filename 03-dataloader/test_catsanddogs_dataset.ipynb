```python
from utils.data import CatsAndDogs
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = CatsAndDogs()
dloader = DataLoader(dataset, batch_size=8, shuffle=True)
```

```python
images, labels = next(iter(dloader))
```

```python
images = images.numpy().transpose(0, 2, 3, 1) + 0.5
labels = labels.numpy()
```

```python
for i in range(8):
    plt.imshow(images[i])
    plt.xticks([], [])
    plt.yticks([], [])
    plt.title(dataset.labels[labels[i]])
    plt.show()
```
